```python
import math
import numpy as np
import matplotlib.pyplot as plt
import time
from datetime import datetime, timedelta
import pandas
from pandas import DataFrame
from pandas_datareader import data as web
from itertools import combinations
# download data from YAHOO FINANCE
#--------------------------------------------------------
def loadYahoo(tickers, dateFrom, dateTo, interval):
all_data = {};
for ticker in tickers:
all_data[ticker] = web.get_data_yahoo(ticker, dateFrom, dateTo, interval=interval);
price = DataFrame({tic: data['Adj Close'] for tic, data in all_data.items()});
priceR = price.pct_change();
ret = {};
for ticker in tickers:
ret[ticker] = price[ticker].tolist()[1:];
retR = {};
for ticker in tickers:
retR[ticker] = priceR[ticker].tolist()[1:];
return(ret,retR);
#--------------------------------------------------------

if __name__ == '__main__':
# ask user to specify tickers
userInput = input("Enter tickers (use space to separate values):");
try:
    
# unpack user input into array
tickers = userInput.upper().split();
if(len(tickers)<3):
raise(Exception("Minimum 3 tickers should be specified"))

# define date range (previous 10 years)
dateStop = datetime.now().strftime("%m/%d/%Y");
dateStart = (datetime.now()-timedelta(days=10*365)).strftime("%m/%d/%Y");

# request data from Yahoo Finance
data, dataR = loadYahoo(tickers, dateStart, dateStop,'m');

# create unique pairs from a list of tickers
comb = list(combinations(tickers,2));

# create areas to plot data (alligned horizontally) 
fig, ax = plt.subplots(nrows=len(comb),ncols=2);

# iterate over each pair
for i in range(len(comb)):
pair = comb[i];
# unpack data into single arrays
x = dataR[pair[0]]; # first ticker
y = dataR[pair[1]]; # second ticker
# calculate correlation coefficient
c = np.corrcoef(x, y)[0, 1];

# plot cummalative returns in 1 area
ax[i][0].plot(np.cumsum(x),label=pair[0]);
ax[i][0].plot(np.cumsum(y),label=pair[1]);
ax[i][0].legend();
# plot returns in 2 area
ax[i][1].scatter(x,y);
ax[i][1].set_title("Correlation: {0:.2f}".format(c));
plt.show();
except Exception as e:
# process exception
print(str(e));
```
