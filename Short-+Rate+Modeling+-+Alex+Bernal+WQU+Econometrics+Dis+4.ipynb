Simulate Interest Rate Path using Vasicek Model

The Vasicek Model   

One-factor model
Short-term interest rate is modeled as a single stochastic factor
Interest rates could be negative according to the model, being an undesirable result
The model takes into consideration mean reversion theory (interest rates move back to their mean)
Interest rates cannot grow indefinitely as compared to stocks

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def vasicek_model(r0, K1, theta1, sigma1, T1=1., N1=10,
                  seed=777):
    np.random.seed(seed)
    dt = T1/float(N1)
    interest_rates = [r0]
    
    for i in range (N1):
        dr = K1*(theta1-interest_rates[-1])*dt +  sigma1*np.random.normal()
        interest_rates.append(interest_rates[-1] + dr)
    return range(N1+1), interest_rates

X1, y1 = vasicek_model(0.025, 0.23, 0.023, 0.01, 10., 200)
```

```python
plt.plot(X1, y1)
```

Cox-Ingersoll-Ross Model (CIR) 

One-factor model
Extension of the Vasicek model
Does not provide negative interest rates as Vasicek model

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cir_model(r0, K1, theta1, sigma1, T1=1., N1=10, seed=777):
    np.random.seed(seed)
    dt = T1/float(N1)
    interest_rates = [r0]
    
    for i in range(N1):
        dr = K1*(theta1-interest_rates[-1])*dt + sigma1*math.sqrt(interest_rates[-1])*np.random.normal()
        interest_rates.append(interest_rates[-1] + dr)
    return range(N1+1), interest_rates

x1, y1 = cir_model(0.024, 0.23, 0.023, 0.01, 10., 200)    
```

```python
plt.plot(x1, y1)
```

The Rendleman and Bartter Model

One-factor model
Does not capture mean reversion
Instantaneous interest rate follows a Geometric Brownian Motion

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rendleman_bartter_model(r0, K1, theta1, sigma1, T1=1., N1=10, seed=777):
    np.random.seed(seed)
    dt = T1/float(N1)
    interest_rates = [r0]
    
    for i in range(N1):
        dr = (theta1*interest_rates[-1])*dt + sigma1*(interest_rates[-1])*np.random.normal()
        interest_rates.append(interest_rates[-1] + dr)
    return range(N1+1), interest_rates

x1, y1 = rendleman_bartter_model(0.025, 0.15, 0.023, 0.01, 10., 200)    
```

```python
plt.plot(x1, y1)
```

The Brennan and Schwartz Model

Two-factor model
It is a form of Geometric Brownian Motion

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def brennan_schwartz_model(r0, K1, theta1, sigma1, T1=1., N1=10, seed=777):
    np.random.seed(seed)
    dt = T1/float(N1)
    interest_rates = [r0]

    for i in range(N1):
        dr = K1*(theta1-interest_rates[-1])*dt + sigma1*(interest_rates[-1])*np.random.normal()
        interest_rates.append(interest_rates[-1] + dr)
    return range(N1+1), interest_rates

x1, y1 = brennan_schwartz_model(0.025, 0.23, 0.023, 0.01, 10., 200) 
```

```python
plt.plot(x1, y1)
```

Vasicek - Negative interest rates are possible
Cox-Ingersoll-Ross - NO negative interest Rates
Rendleman Bartter - Geometric Brownian Motion, log-normally distributed - lacks mean reversion
Brennan Schqartz - Stocastic Methdo where short rate reverts to long rates
